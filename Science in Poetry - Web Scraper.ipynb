```python
def scrape_poems(website_):
    """
    Function scrapes poems from specified website
    """
    response_ = requests.get(website_)
    soup = BeautifulSoup(response_.content, 'html.parser')
    poems_ = soup.find_all('article')
    return poems_
```

```python
def write_poems_to_file(poems, csvfile, category):
    """
    This function write the scraped poems to the csv file
    poems: scraped poems
    csvfile: name of the file
    category: 1 - sci-poems, 0 - nonscipoems
    """
    writer = csv.writer(csvfile)
    for poem in poems:
        title = poem.find('h2').text.strip()
        content = poem.find('div').text.strip()
        writer.writerow([title, content, category])
```

```python

def crawl_website(website, csv_filename, category, page_numbers):
    """
    funtions scrapes multipages in the website by incrementing the page numbers
    """
    with open(csv_filename, 'w', newline='', encoding='utf-8') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(['Title', 'Content', 'Science'])  # header row
        total_count = 0
        for i in range(1, page_numbers+1):
            url_ = f'{website}/page/{i}/'
            print(f'Page URL: {url_}')
            poems_ = scrape_poems(url_)
            print(f'Number of Poems: {len(poems_)}')
            write_poems_to_file(poems_, csvfile, category)
            total_count += len(poems_)
            print(f'{total_count} Poems written to file')
            
```

```python
def scrape_poems2(website_):
    """
    Function scrapes poems dfrom specified website
    """
    response_ = requests.get(website_)
    soup = BeautifulSoup(response_.content, 'html.parser')
    poems_ = soup.find_all('div', class_="sqs-html-content")
    return poems_

def write_poems_to_file2(poem, csvfile, category):
    """
    This function write the scraped poems to the csv file
    poems: scraped poems
    csvfile: name of the file
    category: 1 - sci-poems, 0 - nonscipoems
    """
    writer = csv.writer(csvfile)
#     for poem in poems:
#     print(poem)
    title = poem.find('h1').text.strip()
    content = poem.find('p').text.strip()
    writer.writerow([title, content, category])

    
def write_poems_to_file2b(poem, csvfile, category):
    """
    This function write the scraped poems to the csv file
    poems: scraped poems
    csvfile: name of the file
    category: 1 - sci-poems, 0 - nonscipoems
    """
    writer = csv.writer(csvfile)
#     if len(poem) > 1:
#         poem = poem[0]
    title = poem[0].find('h1').text.strip()
    conterer = poem[0].find_all('p')
    content = ""
    for conter in conterer:
        content +=  f'{conter.text.strip()} '
    writer.writerow([title, content, category])
    

def crawl_website2(websites, csv_filename, category, total_count, indexess):
    """
    funtions scrapes multipages in the website bt incrementing the page numbers
    """
    with open(csv_filename, 'a', newline='', encoding='utf-8') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(['Title', 'Content', 'Science'])  # header row
        for index, i in enumerate(websites):
            if index >= indexess:
                print(f'Page URL: {i}')
                poems_ = scrape_poems2(i)
                write_poems_to_file2b(poems_, csvfile, category)
#             poems_ = scrape_poems(i)
# #             print(f'Number of Poems: {len(poems_)}')
#             write_poems_to_file2(poems_[0], csvfile, category)
                total_count += 1
                print(f'{total_count} Poems written to file')
```

```python
poetsorg = 'https://poets.org/poems?field_occasion_target_id=All&field_poem_themes_target_id=1311&field_form_target_id=All&combine=&page='
poetsorg_links = []

# o - 4, find all table cols-3, find a, href

for i in range(5):
    org = f'{poetsorg}{i}'
    print(f'website is - {org}')
    resp_ = requests.get(org)
    soup = BeautifulSoup(resp_.content, 'html.parser')
    tables = soup.find_all('td', class_="views-field views-field-title")
    for table in tables:
#         print(table)
        linka = table.find('a')
#         print('let us find the anchors')
#         print(linka.get('href'))
        poetsorg_links.append(f"https://poets.org{linka.get('href')}")
        
```
